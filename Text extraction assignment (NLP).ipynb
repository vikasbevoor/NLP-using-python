## Text extraction

```python
# Extracting the speech content from the page
```

```python
Extract_text =[]
```

```python
content = browser.page_source
soup = bs(content, "html.parser")
rev = soup.findAll("p", attrs={"class","css-158dogj evys1bk0"})
blacklist = [
    '[document]',
    'noscript',
    'header',
    'html',
    'meta',
    'head', 
    'input',
    'script'
]

for t in rev:
    if t.parent.name not in blacklist:
        output = rev
text = [e.get_text() for e in output]
Extract_text.extend(text)
```

```python
# Extracted text from the page
```

```python
Extract_text
```

```python
len(Extract_text)
```

#### There are totally 43 paragrapghs in the speech

## Text processing

```python
# Removing punctuations, numbers
```

```python
clean_text = re.sub("[^A-Za-z0-9" "]+"," ",str(Extract_text))
clean_text = clean_text.lower()
clean_text
```

```python
# Saving the speech words in the form of text file
```

```python
with open("Text_DS.txt", "w", encoding="utf8") as output:
    output.write(str(Extract_text))
```

```python
# Splitting each word of the speech
```

```python
text_words = clean_text.split(" ")
text_words
```

```python
len(text_words)
```

```python
# Importing the stop words
```

```python
with open(r"D:\Data science\Python codes\Data sets\NLP words list\stopwords.txt","r") as sw:
    stopwords = sw.read()
```

```python
stopwords = stopwords.split("\n")
```

```python
# Removing the stop words from the speech
```

```python
text_actual = [w for w in text_words if not w in stopwords]
text_actual
```

```python
len(text_actual)
```

```python
# Joining all the words into a single paragraph
```

```python
clean_text = " ".join(text_actual)
clean_text
```
