# Amazon review sentimental analysis

```python
# Importing necessary libraries
```

```python
import requests 
import re
from bs4 import BeautifulSoup as bs
```

## Reviews extraction

```python
# Using the amazon reviews of "One plus phone"
```

```python
oneplus_reviews=[]
```

```python
for i in range(1,6):
    op=[]
    url="https://www.amazon.in/Test-Exclusive-748/product-reviews/B07DJLVJ5M/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber="+str(i)
    response = requests.get(url)
    soup = bs(response.content, "html.parser")
    reviews = soup.findAll("span", attrs={"class":"a-size-base review-text review-text-content"})
    for j in range(len(reviews)):
        op.append(reviews[j].text)
    oneplus_reviews = oneplus_reviews+op
    
```

```python
# Checking the reviews obtained from amazon website
```

```python
oneplus_reviews
```

```python
len(oneplus_reviews)                # No. of paragraphs in the reviews
```

```python
# Saving the reviews in text format
```

```python
with open("oneplus_reviews.txt","w", encoding="utf8") as output:
    output.write(str(oneplus_reviews))
```

```python
# Joining all the reviews into one paragraph
```

```python
op_rev_string = " ".join(oneplus_reviews)
op_rev_string
```

```python
len(op_rev_string)             # No. of words in the paragraph
```

## Text processing

```python
# Removing the punctuation marks and numbers from the words and converting all the words into lower case
```

```python
op_rev_string= re.sub("[^A-Za-z" "]+"," ", op_rev_string).lower()
```

```python
op_rev_string=re.sub("[0-9" "]+"," ", op_rev_string)
op_rev_string
```

```python
# Splitting each word from the paragraph
```

```python
op_rev_words = op_rev_string.split(" ")
op_rev_words
```

```python
len(op_rev_words)
```

#### There are total 5933 words in the reviews paragraph

```python
# Importing the stop words
```

```python
with open(r"D:\Data science\Python codes\Data sets\NLP words list\stopwords.txt","r") as sw:
    stopwords = sw.read()
```

```python
stopwords= stopwords.split("\n")
```

```python
stopwords=stopwords[6:]
stopwords
```

```python
# Removing all the stopwords from the review words
```

```python
op_rev_words = [w for w in op_rev_words if not w in stopwords]
op_rev_words
```

```python
op_rev_string = " ".join(op_rev_words)
op_rev_string
```
